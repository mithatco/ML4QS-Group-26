### Get list of timestamps

```python
import requests
import pandas as pd
from datetime import datetime

# Read the CSV file
df = pd.read_csv('Heart Rate 2.csv')

# Extract the timestamps column
timestamps = df['Date/Time']

# Convert the timestamps to datetime objects
timestamps = pd.to_datetime(timestamps)

# Function to determine location based on month
def determine_location(timestamp):
    if timestamp.month in [7, 8]:
        return 'Turkey'
    else:
        return 'Amsterdam'

# Apply the function to create a new 'Location' column
df['Location'] = timestamps.apply(determine_location)

# Define your API key and base URL
api_key = '1d42a91aa396e88749ea11b7439adca9'
base_url = 'http://api.openweathermap.org/data/2.5/onecall/timemachine'

location_amsterdam = {'lat': 52.3676, 'lon': 4.9041}
location_turkey = {'lat': 38.4237, 'lon': 27.1428}

# Convert timestamps to UNIX format
unix_times = [int(ts.timestamp()) for ts in timestamps]

# Fetch weather data for each timestamp
weather_data = []
# Process and store the weather data
weather_df = pd.DataFrame(weather_data)

for unix_time, location in zip(unix_times, df['Location']):
    if location == 'Amsterdam':
        coords = location_amsterdam
    else:
        coords = location_turkey
        
    response = requests.get(base_url, params={
        'lat': coords['lat'],
        'lon': coords['lon'],
        'dt': unix_time,
        'appid': api_key
    })
    data = response.json()
    print(data)
    weather_data.append(data)
    weather_df = pd.DataFrame(weather_data)
    weather_df.to_csv('weather_data.csv', index=False)

# Process and store the weather data
weather_df = pd.DataFrame(weather_data)

#save the weather data to a csv file
weather_df.to_csv('weather_data.csv', index=False)
```
